***1) Description of the data set:***

This dataset presents real estate information and various features of each house. Each row is a representation of specific properties, while the columns are:

X1 transaction date - in terms of (year and month).

X2 age of the house - age in annual terms.

X3 distance to the nearest MRT station - measured in meters.

X4 number of convenience stores - number.

X5 latitude - coordinates of the house location.

X6 longitude- coordinates of the house location.

Y price of a house with a unit area - price per unit of house area.

***2) A statistical description of the first 100 entries of the data set:***

```python
import pandas as pd
import io

url = 'https://raw.githubusercontent.com/Jaciubek/goitneo-math-hw-10/main/Realestate.csv'
df = pd.read_csv(url)
df.head()

#To see how to fully display the tables correctly, click on this link https://colab.research.google.com/notebooks/data_table.ipynb#scrollTo=3jcHW3nRJpaE
```

```python
# creating statistical description of the first 100 rows:

first_100_rows = df.head(100)
first_100_rows.describe()
#To see how to fully display the tables correctly, click on this link https://colab.research.google.com/notebooks/data_table.ipynb#scrollTo=3jcHW3nRJpaE
```

***3) The forecasting goal - price:***

```python
y =df["Y house price of unit area"]
y.head()
```

***4) Using available parameters for forecasting:***



```python
# printing column names:

df.columns
```

```python
# selection of parameters for prediction:

from sklearn.model_selection import train_test_split

realestate_features = ["X1 transaction date", "X2 house age", "X3 distance to the nearest MRT station", "X4 number of convenience stores", "X5 latitude", "X6 longitude" ]

X = df[realestate_features]

# division(split) of data into training and testing sets:

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

X_train.head()
```

***5) Creating a model using DecisionTreeRegressor***

```python
from sklearn.tree import DecisionTreeRegressor

# create and train the model:

realestate_model = DecisionTreeRegressor(random_state=1)
realestate_model.fit(X_train, y_train)
```

Model has been created

***6) Property price prediction for the first 5 rows:***

```python
first_5_rows = X.head(5)
predictions = realestate_model.predict(first_5_rows)

predictions
```

***7) Create a new model and predict the house price using only the house coordinates:***

```python
# select features (only latitude and longitude):

realestate_features_b = ["X5 latitude", "X6 longitude"]

X_coords = df[realestate_features_b]

X_coords.head()
```

```python
# division(split) of data into training and testing sets:

X_train_coords, X_test_coords, y_train_coords, y_test_coords = train_test_split(X_coords, y, test_size=0.2, random_state=1)

# create and train the model:

model_coords = DecisionTreeRegressor(random_state=1)
model_coords.fit(X_train_coords, y_train_coords)
```

Model has been created

```python
# Make a prediction for the first 5 rows using only coordinates:

first_5_rows_coords = X_coords.head(5)
predictions_coords = model_coords.predict(first_5_rows_coords)

predictions_coords
```

***8) Conclusions***

```python
# forecasts(predictions) from both models:
print("Forecasts with using all features:", predictions)
print("Forecasts with using only coordinates:", predictions_coords)

# reference to the current prices:
current_prices = y.head(5).values
print("Current prices:", current_prices)

# difference analysis:
differences_all_features = current_prices - predictions
differences_coords = current_prices - predictions_coords

print("Differences with all features:", differences_all_features)
print("Differences with coordinates only:", differences_coords)
```

**A model based on multiple parameters provides greater accuracy in predicting house prices than a model based solely on geographic coordinates.**
